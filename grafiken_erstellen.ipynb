```python
import numpy as np

# Paket zum Erstellen von Grafiken (Plotten)
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-1, 1)
y = x ** 2 #+ np.random.randn(len(x)) * 0.3
dy = 2 * x
```

```python
#plt.figure(figsize=(15, 10))
plt.plot(x, y, 'g') # Funktion
plt.plot(x, dy, 'r--') # Ableitung
plt.xlabel('x-Achse')
plt.ylabel('y-Achse')
plt.title('Funktion und ihre Ableitung')
plt.xlim([-0.1, 0.1])
plt.ylim([-0.25, 0.25])
plt.legend(['Funktion', 'Ableitung'])
plt.show()

plt.plot(x, dy)
plt.xlabel('x-Achse')
plt.ylabel('y-Achse')
plt.title('Ableitung')

plt.show()
```
